```python
data = np.genfromtxt("C:/Users/kabdullah/Top 10.csv", delimiter=',',dtype=int)
# Choose how many bins you want here
num_bins = 86
# Use the histogram function to bin the data
counts, bin_edges = np.histogram(data, bins=num_bins, normed=True)
# Now find the cdf
cdf = np.cumsum(counts)
# And finally plot the cdf
plt.figure()
plt.plot(bin_edges[1:], cdf,linewidth=2.5 , ls='solid' , animated='True')
plt.ylim((0,1.02))
plt.grid(True, linestyle='--', linewidth=1, which='major')
plt.legend(['High MCS Penetration for the top 10% Busy Cells'], loc='best')
#plt.title("CDF for the top 10% Busy Cells")
plt.xlabel("High MCS Penetration(%)")
plt.ylabel("Cummulative distribution", fontsize='large')
savefig('foo.pdf', bbox_inches='tight')
plt.show()
```
