```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import savefig
from matplotlib.backends.backend_pdf import PdfPages
from pylab import *
from matplotlib import rc

ser=np.genfromtxt("C:/Users/kabdullah/Paper/MCS Enhancement.csv", delimiter=',',dtype=float)
fig = plt.figure()
ax1 = fig.add_subplot(111)
#2nd Axis
lns1=ax1.plot(ser[:,0], ser[:,1],linewidth=3.5 , ls='solid' ,marker ="X",
              label="$\\bf{High MCS Pentration (\%)}$")
ax1.set_xlabel("$\\bf{Monitored Days}$", fontsize='xx-large')
ax1.set_ylabel("$\\bf{High MCS Pentration (\%) }$", fontsize='xx-large')
plt.ylim((60,80))
#Second Axis
ax2 = ax1.twinx()
lns2=ax2.plot(ser[:,0], ser[:,2],linewidth=3.5 , ls='solid' , color='black',
              marker ="o",label="$\\bf{Average UL UE Throughput}$")
ax2.set_ylabel("$\\bf{UL User Throughput (Mbps)}$", fontsize='xx-large')
plt.ylim((1,2))
# To adjust the ticks
ax1.set_xticks(ser[:,0], minor=True)
#Grid
ax1.grid(True,linestyle='--',linewidth=1,which='both', alpha=0.3)
#plt.grid(True, linestyle='--', linewidth=1, which='major', fillstyle='full')
# Legends
lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc='lower right',fontsize='x-large',framealpha=.2)
#Title
#plt.title("Uplink User Throughput Enhancement after MCS Adjustment")
plt.rcParams["figure.figsize"] = (10,4)
#SaveAsPDF
savefig('PhysicalOptResult.pdf', bbox_inches='tight')
plt.show()
```
